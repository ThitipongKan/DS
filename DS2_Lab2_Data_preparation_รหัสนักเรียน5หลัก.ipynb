- ‡∏ù‡∏∂‡∏Å‡πÉ‡∏ä‡πâ `merge()` ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏≤‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏Å‡∏±‡∏ô


```python
import pandas as pd
```

## üîπ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÉ‡∏ô‡∏´‡πâ‡∏≠‡∏á A
data_a = pd.DataFrame({
    'student_id': [1, 2, 3],
    'name': ['Ann', 'Bob', 'Cat'],
    'score': [85, 78, 92]
})

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÉ‡∏ô‡∏´‡πâ‡∏≠‡∏á B
data_b = pd.DataFrame({
    'student_id': [4, 5, 6],
    'name': ['Dan', 'Eve', 'Fin'],
    'score': [88, 90, 75]
})

print("üìò ‡∏´‡πâ‡∏≠‡∏á A")
print(data_a)
print("\nüìó ‡∏´‡πâ‡∏≠‡∏á B")
print(data_b)
```

## üîπ ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡πâ‡∏ß‡∏¢ `concat()`

pd.concat([data_a, data_b], ...)

* pd.concat() ‡∏Ñ‡∏∑‡∏≠‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÉ‡∏ô pandas ‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ ‡∏£‡∏ß‡∏°‡∏ï‡∏≤‡∏£‡∏≤‡∏á (DataFrame) ‡∏´‡∏•‡∏≤‡∏¢ ‡πÜ ‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡πÄ‡∏Ç‡πâ‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏±‡∏ô

* ‡πÉ‡∏ô‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö [data_a, data_b] ‡∏´‡∏°‡∏≤‡∏¢‡∏ñ‡∏∂‡∏á‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏≤ ‡∏ï‡∏≤‡∏£‡∏≤‡∏á data_a ‡πÅ‡∏•‡∏∞ data_b ‡∏°‡∏≤‡∏ï‡πà‡∏≠‡∏Å‡∏±‡∏ô

* ‡∏£‡∏∞‡∏ö‡∏∏ ‚Äú‡πÅ‡∏ô‡∏ß‚Äù ‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•

    * axis=0 ‚Üí ‡∏£‡∏ß‡∏° ‡πÅ‡∏ô‡∏ß‡∏ï‡∏±‡πâ‡∏á (‡∏ï‡πà‡∏≠‡πÅ‡∏ñ‡∏ß) ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏≠‡∏≤‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡∏°‡∏≤‡∏ï‡πà‡∏≠‡πÉ‡∏ï‡πâ‡∏Å‡∏±‡∏ô

    * axis=1 ‚Üí ‡∏£‡∏ß‡∏° ‡πÅ‡∏ô‡∏ß‡∏ô‡∏≠‡∏ô (‡∏ï‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå) ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏≠‡∏≤‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡∏°‡∏≤‡∏ï‡πà‡∏≠‡∏Ç‡πâ‡∏≤‡∏á ‡πÜ ‡∏Å‡∏±‡∏ô

* ‡∏õ‡∏Å‡∏ï‡∏¥‡πÄ‡∏°‡∏∑‡πà‡∏≠ concat ‡πÅ‡∏•‡πâ‡∏ß pandas ‡∏à‡∏∞‡∏¢‡∏±‡∏á ‡πÄ‡∏Å‡πá‡∏ö index ‡πÄ‡∏î‡∏¥‡∏° ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡πÑ‡∏ß‡πâ ‡∏ñ‡πâ‡∏≤‡πÉ‡∏™‡πà ignore_index=True ‡∏à‡∏∞‡πÉ‡∏´‡πâ pandas ‡∏™‡∏£‡πâ‡∏≤‡∏á index ‡πÉ‡∏´‡∏°‡πà‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà 0, 1, 2, ... ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏≤‡∏°‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î

```python
all_students = pd.concat([data_a, data_b], axis=0, ignore_index=True)
print("üìö ‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏´‡πâ‡∏≠‡∏á:")
print(all_students)
```

###  ‡∏£‡∏ß‡∏°‡πÅ‡∏ô‡∏ß‡∏ï‡∏±‡πâ‡∏á (concat columns)

```python
grade_data = pd.DataFrame({
    'grade': ['‡∏°.4', '‡∏°.4', '‡∏°.4', '‡∏°.5', '‡∏°.5', '‡∏°.5']
})

data1 = pd.concat([all_students, grade_data], axis=1)
print("üìä ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢:")
print(data1)
```

## üîπ ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡πâ‡∏ß‡∏¢ `merge()`

‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ 4 ‡∏ä‡∏∏‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏Å‡∏±‡∏ô

*   restaurants ‚Äî ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£ (restaurant_id, name, cuisine, province)
*   menu_items ‚Äî ‡πÄ‡∏°‡∏ô‡∏π‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏£‡πâ‡∏≤‡∏ô (menu_id, restaurant_id, item_name, price)
*   orders ‚Äî ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Å‡∏≤‡∏£‡∏™‡∏±‡πà‡∏á‡∏≠‡∏≤‡∏´‡∏≤‡∏£ (order_id, restaurant_id, menu_id, quantity, datetime)
*   suppliers ‚Äî ‡∏ú‡∏π‡πâ‡∏à‡∏±‡∏î‡∏™‡πà‡∏á‡∏ß‡∏±‡∏ï‡∏ñ‡∏∏‡∏î‡∏¥‡∏ö‡πÉ‡∏´‡πâ‡∏£‡πâ‡∏≤‡∏ô (supplier_id, restaurant_id, supplier_name, ingredient_type)









```python
import pandas as pd
from IPython.display import display
import numpy as np

# restaurants
restaurants = pd.DataFrame({
    'restaurant_id': [201, 202, 203, 204, 205],
    'name': ['‡∏ö‡πâ‡∏≤‡∏ô‡∏Å‡∏•‡∏≤‡∏á‡∏Å‡∏£‡∏∏‡∏á', '‡∏Ç‡πâ‡∏≤‡∏ß‡∏°‡∏±‡∏ô‡πÑ‡∏Å‡πà‡∏Ñ‡∏∏‡∏ì‡πÄ‡∏•‡πá‡∏Å', '‡∏™‡πâ‡∏°‡∏ï‡∏≥‡∏ã‡∏≠‡∏¢ 9', '‡∏£‡πâ‡∏≤‡∏ô‡∏û‡∏¥‡∏ã‡∏ã‡πà‡∏≤‡πÇ‡∏Å‡πâ', '‡∏Ç‡∏ô‡∏°‡∏à‡∏µ‡∏ô‡πÅ‡∏°‡πà‡∏ö‡∏∏‡∏ç'] ,
    'cuisine': ['Thai', 'Thai', 'Isaan', 'Italian', 'Thai'],
    'province': ['Bangkok', 'Chiang Mai', 'Khon Kaen', 'Bangkok', 'Phuket']
})

# menu_items
menu_items = pd.DataFrame({
    'menu_id': ['M1','M2','M3','M4','M5','M6','M7'],
    #'restaurant_id': [201,201,202,203,204,204,206],  # note: 206 is an orphan menu (restaurant not in restaurants)
    'item_name': ['‡∏ú‡∏±‡∏î‡πÑ‡∏ó‡∏¢','‡∏ï‡πâ‡∏°‡∏¢‡∏≥‡∏Å‡∏∏‡πâ‡∏á','‡∏Ç‡πâ‡∏≤‡∏ß‡∏°‡∏±‡∏ô‡πÑ‡∏Å‡πà','‡∏™‡πâ‡∏°‡∏ï‡∏≥','‡∏°‡∏≤‡∏£‡πå‡πÄ‡∏Å‡∏£‡∏¥‡∏ï‡πâ‡∏≤','‡πÄ‡∏õ‡∏õ‡πÄ‡∏õ‡∏≠‡πÇ‡∏£‡∏ô‡∏µ','‡∏•‡∏≤‡∏ã‡∏≤‡∏ô‡∏ç‡πà‡∏≤'],
    'price': [80,120,60,50,250,270,300]
})

# orders
orders = pd.DataFrame({
    'order_id': [1001,1002,1003,1004,1005,1006,1007,1008],
    'restaurant_id': [201,201,202,203,204,204,206,206],
    'menu_id': ['M1','M2','M3','M4','M5','M5','M7', 'M8'],
    'quantity': [2,1,3,2,1,4,1,3],
    'datetime': pd.to_datetime(['2025-10-01 11:05','2025-10-01 12:30','2025-10-02 18:20','2025-10-03 13:15','2025-10-04 20:00','2025-10-05 19:45','2025-10-05 12:00','2025-10-05 16:00'])
})

# suppliers
suppliers = pd.DataFrame({
    'supplier_id': [301,302,303,304],
    'restaurant_id': [201,202,204,207],  # 207 is a supplier connection to a restaurant not in restaurants
    'supplier_name': ['‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó‡∏ú‡∏±‡∏Å‡∏™‡∏î','‡∏ü‡∏≤‡∏£‡πå‡∏°‡πÑ‡∏Å‡πà‡∏™‡∏°‡∏õ‡∏£‡∏∞‡∏™‡∏¥‡∏ó‡∏ò‡∏¥‡πå','‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó‡πÅ‡∏õ‡πâ‡∏á‡∏û‡∏¥‡∏ã‡∏ã‡πà‡∏≤','‡∏ã‡∏±‡∏û‡∏û‡∏•‡∏≤‡∏¢‡πÄ‡∏≠‡∏≠‡∏£‡πå‡∏Ç‡∏ô‡∏°‡∏à‡∏µ‡∏ô'],
    'ingredient_type': ['vegetable','chicken','flour','rice']
})

print('restaurants')
display(restaurants)
print('menu_items')
display(menu_items)
print('orders')
display(orders)
print('suppliers')
display(suppliers)
```

‡∏à‡∏∞‡πÄ‡∏´‡πá‡∏ô‡∏ß‡πà‡∏≤‡∏ä‡∏∏‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á 4 ‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡∏µ‡πà‡∏¢‡∏ß‡∏Ç‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ô ‡πÅ‡∏ï‡πà‡∏ñ‡∏π‡∏Å‡πÄ‡∏Å‡πá‡∏ö‡πÅ‡∏¢‡∏Å‡∏Å‡∏±‡∏ô‡πÑ‡∏ß‡πâ ‡πÇ‡∏î‡∏¢‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏≠‡∏ö‡∏Ñ‡∏≥‡∏ñ‡∏≤‡∏°‡∏ö‡∏≤‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡∏ä‡∏∏‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 1 ‡∏ä‡∏∏‡∏î ‡πÄ‡∏ä‡πà‡∏ô ‡∏´‡∏≤‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏£‡∏≤‡∏ö‡∏ß‡πà‡∏≤‡∏£‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏£‡πâ‡∏≤‡∏ô‡πÉ‡∏î‡∏Ç‡∏≤‡∏¢‡∏ú‡∏±‡∏î‡πÑ‡∏ó ‡∏à‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡∏ó‡∏±‡πâ‡∏á menu_items, orders ‡πÅ‡∏•‡∏∞ restaurants ‡∏õ‡∏£‡∏∞‡∏Å‡∏≠‡∏ö‡∏Å‡∏±‡∏ô ‡∏ö‡∏≤‡∏á‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÄ‡∏£‡∏≤‡∏à‡∏∂‡∏á‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏ä‡∏∏‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏Ç‡πâ‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏±‡∏ô ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á merge ‡∏ã‡∏∂‡πà‡∏á‡∏°‡∏µ‡∏ß‡∏¥‡∏ò‡∏µ‡∏Å‡∏≤‡∏£‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î 4 ‡∏ß‡∏¥‡∏ò‡∏µ

‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô merge() ‡∏Ç‡∏≠‡∏á pandas ‡πÉ‡∏ä‡πâ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö ‚Äú‡∏£‡∏ß‡∏°‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡∏ï‡∏≤‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏±‡∏ô (join)‚Äù
‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏≥ JOIN ‡πÉ‡∏ô‡∏ê‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• SQL ‡πÄ‡∏ä‡πà‡∏ô INNER JOIN, LEFT JOIN, RIGHT JOIN ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πâ‡∏ô ‡πÇ‡∏î‡∏¢ ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á merge() :
pd.merge(left, right, on='key', how='...')

* on : ‡∏£‡∏∞‡∏ö‡∏∏‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏∏‡∏ç‡πÅ‡∏à (key) ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö join (‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô‡πÉ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏ï‡∏≤‡∏£‡∏≤‡∏á)

* how = ‡∏ö‡∏≠‡∏Å pandas ‡∏ß‡πà‡∏≤‡πÄ‡∏ß‡∏•‡∏≤ merge ‡∏™‡∏≠‡∏á‡∏ï‡∏≤‡∏£‡∏≤‡∏á
‡∏à‡∏∞ ‚Äú‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ù‡∏±‡πà‡∏á‡πÑ‡∏´‡∏ô‡∏ö‡πâ‡∏≤‡∏á‚Äù ‡∏ñ‡πâ‡∏≤‡∏Ñ‡∏µ‡∏¢‡πå‡πÑ‡∏°‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô
    * | ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á | ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡∏ù‡∏±‡πà‡∏á‡πÑ‡∏´‡∏ô       |
| ------ | -------------------------- |
| inner  | ‡πÄ‡∏â‡∏û‡∏≤‡∏∞ key ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏Ñ‡∏π‡πà |
| left   | ‡πÄ‡∏Å‡πá‡∏ö‡∏ù‡∏±‡πà‡∏á‡∏ã‡πâ‡∏≤‡∏¢‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î        |
| right  | ‡πÄ‡∏Å‡πá‡∏ö‡∏ù‡∏±‡πà‡∏á‡∏Ç‡∏ß‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î         |
| outer  | ‡πÄ‡∏Å‡πá‡∏ö‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏ù‡∏±‡πà‡∏á            |


```python
# Example 1: inner join between menu_items and orders on menu_id (shows orders that have matching menu)
inner_m_o = pd.merge(menu_items, orders, on='menu_id', how='inner')
print('Inner join (menu_items ‚®ù orders)')
display(inner_m_o)
```

```python
# Example 2: left join between menu_items and orders on menu_id (shows all menu items)
left_m_o = pd.merge(menu_items, orders, on='menu_id', how='left')
print('Left join (menu_items ‚®ù orders)')
display(left_m_o)
```

```python
# Example 3: right join between menu_items and orders on menu_id (shows all orders)
right_m_o = pd.merge(menu_items, orders, on='menu_id', how='right')
print('Right join (menu_items ‚®ù orders)')
display(right_m_o)
```

```python
# Example 4: outer join between menu_items and orders on menu_id (shows everything)
outer_m_o = pd.merge(menu_items, orders, on='menu_id', how='outer')
print('Outer join (menu_items ‚®ù orders)')
display(outer_m_o)
```
